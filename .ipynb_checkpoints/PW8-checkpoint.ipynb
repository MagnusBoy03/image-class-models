```python
class_names = sorted(df["breed_name"].unique())
class_to_idx = {name: i for i, name in enumerate(class_names)}

df["label"] = df["breed_name"].map(class_to_idx)

df[["breed_name", "label"]].head(10)
```

```python
from torch.utils.data import Dataset

class PetDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.df = dataframe
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        
        image = Image.open(row["image_path"]).convert("RGB")
        label = row["label"]
        
        if self.transform:
            image = self.transform(image)
        
        return image, label
```

```python
dataset = PetDataset(df, transform=basic_transform)

img, label = dataset[0]

print("Image shape:", img.shape)
print("Label:", label)
```

### 1.2.2. Split Dataset

```python
from sklearn.model_selection import train_test_split

train_df, temp_df = train_test_split(
    df,
    test_size=0.30,
    stratify=df["label"],
    random_state=42
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    stratify=temp_df["label"],
    random_state=42
)

print("Train:", len(train_df))
print("Validation:", len(val_df))
print("Test:", len(test_df))
```

The dataset was divided into training, validation, and test sets to support proper model development and evaluation. The training set was used to learn model parameters, the validation set was used for tuning and monitoring performance, and the test set was reserved for final unbiased evaluation.

```python
train_dataset = PetDataset(train_df, transform=basic_transform)
val_dataset = PetDataset(val_df, transform=basic_transform)
test_dataset = PetDataset(test_df, transform=basic_transform)

print("Train dataset size:", len(train_dataset))
print("Validation dataset size:", len(val_dataset))
print("Test dataset size:", len(test_dataset))
```

```python
from torch.utils.data import DataLoader

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
images, labels = next(iter(train_loader))

print("Batch image shape:", images.shape)
print("Batch labels shape:", labels.shape)
print("First 10 labels:", labels[:10])
```

# 2. Experimental Design

This section presents the model development process used to compare different deep learning approaches for image classification. A baseline CNN and multiple pretrained architectures were trained and evaluated to examine differences in performance, efficiency and generalization.

## 2.1 Baseline CNN

A baseline CNN was first implemented to provide a simple reference point for later experiments. This model makes it easier to assess whether pretrained and more advanced architectures offer meaningful improvements in classification performance

```python
import torch.nn as nn

class BaselineCNN(nn.Module):
    def __init__(self, num_classes=37):
        super(BaselineCNN, self).__init__()
        
        self.features = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128 * 28 * 28, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x
```

```python
model = BaselineCNN(num_classes=num_classes)

sample_batch, sample_labels = next(iter(train_loader))
outputs = model(sample_batch)

print("Input shape:", sample_batch.shape)
print("Output shape:", outputs.shape)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
loss = criterion(outputs, sample_labels)
print("Loss on sample batch:", loss.item())
```

```python
model.train()

running_loss = 0.0

for images, labels in train_loader:
    optimizer.zero_grad()          # clear old gradients
    outputs = model(images)        # forward pass
    loss = criterion(outputs, labels)  # calculate loss
    loss.backward()                # calculate gradients
    optimizer.step()               # update weights
    
    running_loss += loss.item()

epoch_loss = running_loss / len(train_loader)
print("Training loss:", round(epoch_loss, 4))
```

```python
model.eval()

correct = 0
total = 0

with torch.no_grad():
    for images, labels in val_loader:
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

val_accuracy = correct / total
print("Validation accuracy:", round(val_accuracy, 4))
```

```python
num_epochs = 5

for epoch in range(num_epochs):
    # TRAIN
    model.train()
    running_loss = 0.0

    for images, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    train_loss = running_loss / len(train_loader)

    # VALIDATION
    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in val_loader:
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)

            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    val_accuracy = correct / total

    print(f"Epoch {epoch+1}/{num_epochs} - Loss: {train_loss:.4f} - Val Acc: {val_accuracy:.4f}")
```

The baseline CNN required approximately 20–25 minutes for 5 epochs of training on the available hardware

The baseline CNN showed gradual learning, with training loss decreasing from about 3.42 to 2.17 over 5 epochs, but validation accuracy remained low at around 13.5%. This suggests that while the model was able to learn some useful patterns, its overall classification performance was limited, making it an appropriate benchmark to improve upon with pretrained architectures.

```python
from sklearn.metrics import accuracy_score, f1_score
import numpy as np

model.eval()

all_preds = []
all_labels = []

with torch.no_grad():
    for images, labels in val_loader:
        outputs = model(images)
        _, preds = torch.max(outputs, 1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

baseline_acc = accuracy_score(all_labels, all_preds)
baseline_f1 = f1_score(all_labels, all_preds, average="macro")

print("Baseline Accuracy:", baseline_acc)
print("Baseline F1-score:", baseline_f1)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import label_binarize
import torch.nn.functional as F

model.eval()

all_probs = []
all_labels = []

with torch.no_grad():
    for images, labels in val_loader:
        outputs = model(images)
        probs = F.softmax(outputs, dim=1)

        all_probs.extend(probs.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

all_probs = np.array(all_probs)
all_labels = np.array(all_labels)

all_labels_bin = label_binarize(all_labels, classes=np.arange(num_classes))

baseline_auc = roc_auc_score(all_labels_bin, all_probs, multi_class="ovr", average="macro")

print("Baseline AUC:", baseline_auc)
```

The ROC-AUC score was higher than the accuracy and F1-score, indicating that the model captured some class-separation ability, but its final predictions were still weak in practical classification terms.

## 2.2. ResNet18

ResNet18 was used as a pretrained architecture to test whether transfer learning could improve performance over the baseline CNN. By reusing features learned from a large-scale dataset, the model can capture more informative visual patterns with less training from scratch.

```python
from torchvision import models

resnet18 = models.resnet18(pretrained=True)

# freeze all layers
for param in resnet18.parameters():
    param.requires_grad = False

# replace final layer
resnet18.fc = nn.Linear(resnet18.fc.in_features, num_classes)
```

```python
criterion = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(resnet18.fc.parameters(), lr=0.001)
```

```python
import time

num_epochs = 2

start_time = time.time()

for epoch in range(num_epochs):
    # TRAIN
    resnet18.train()
    running_loss = 0.0

    for images, labels in train_loader:
        optimizer.zero_grad()
        outputs = resnet18(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    train_loss = running_loss / len(train_loader)

    # VALIDATION
    resnet18.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in val_loader:
            outputs = resnet18(images)
            _, predicted = torch.max(outputs, 1)

            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    val_acc = correct / total

    print(f"Epoch {epoch+1}/{num_epochs} - Loss: {train_loss:.4f} - Val Acc: {val_acc:.4f}")

end_time = time.time()

print("Training time:", round(end_time - start_time, 2), "seconds")
```

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
from sklearn.preprocessing import label_binarize
import torch.nn.functional as F
import numpy as np

resnet18.eval()

all_preds = []
all_labels = []
all_probs = []

with torch.no_grad():
    for images, labels in val_loader:
        outputs = resnet18(images)
        probs = F.softmax(outputs, dim=1)
        _, preds = torch.max(outputs, 1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())
        all_probs.extend(probs.cpu().numpy())

all_preds = np.array(all_preds)
all_labels = np.array(all_labels)
all_probs = np.array(all_probs)

# Accuracy
acc = accuracy_score(all_labels, all_preds)

# F1-score
f1 = f1_score(all_labels, all_preds, average="macro")

# AUC
all_labels_bin = label_binarize(all_labels, classes=np.arange(num_classes))
auc = roc_auc_score(all_labels_bin, all_probs, multi_class="ovr", average="macro")

print("ResNet18 Accuracy:", acc)
print("ResNet18 F1-score:", f1)
print("ResNet18 AUC:", auc)
```

ResNet18 performed substantially better than the baseline CNN, showing that transfer learning was much more effective for this classification task. The higher validation performance indicates that pretrained deep features were better able to capture the visual differences between classes.

## 2.3. Fine tuning

After evaluating ResNet18 as a transfer learning model, fine-tuning was performed to allow more of the pretrained layers to adapt to the target dataset. This helps the model learn task-specific features instead of relying only on the original pretrained representations.

```python
for param in resnet18.parameters():
    param.requires_grad = False

for param in resnet18.layer4.parameters():
    param.requires_grad = True

for param in resnet18.fc.parameters():
    param.requires_grad = True
```

```python
optimizer = torch.optim.Adam(filter(lambda p: p.requires_grad, resnet18.parameters()), lr=0.0001)
```

```python
import time

num_epochs = 2  # keep small first

start_time = time.time()

for epoch in range(num_epochs):
    # TRAIN
    resnet18.train()
    running_loss = 0.0

    for images, labels in train_loader:
        optimizer.zero_grad()
        outputs = resnet18(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    train_loss = running_loss / len(train_loader)

    # VALIDATION
    resnet18.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in val_loader:
            outputs = resnet18(images)
            _, predicted = torch.max(outputs, 1)

            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    val_acc = correct / total

    print(f"[Fine-tune] Epoch {epoch+1}/{num_epochs} - Loss: {train_loss:.4f} - Val Acc: {val_acc:.4f}")

end_time = time.time()

print("Fine-tuning time:", round(end_time - start_time, 2), "seconds")
```

Fine-tuning further improved the performance of ResNet18, showing that adjusting deeper layers helped the model better match the characteristics of the dataset. This suggests that the task benefited not only from pretrained features, but also from additional adaptation to the target classes.

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
from sklearn.preprocessing import label_binarize
import torch.nn.functional as F
import numpy as np

resnet18.eval()

all_preds = []
all_labels = []
all_probs = []

with torch.no_grad():
    for images, labels in val_loader:
        outputs = resnet18(images)
        probs = F.softmax(outputs, dim=1)
        _, preds = torch.max(outputs, 1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())
        all_probs.extend(probs.cpu().numpy())

all_preds = np.array(all_preds)
all_labels = np.array(all_labels)
all_probs = np.array(all_probs)

# Accuracy
acc = accuracy_score(all_labels, all_preds)

# F1-score
f1 = f1_score(all_labels, all_preds, average="macro")

# AUC
all_labels_bin = label_binarize(all_labels, classes=np.arange(num_classes))
auc = roc_auc_score(all_labels_bin, all_probs, multi_class="ovr", average="macro")

print("FIned-Tuned Accuracy:", acc)
print("Fine-Tuned F1-score:", f1)
print("Fine-Tuned AUC:", auc)
```

| Model                 | Accuracy   | F1-score   | AUC        | Training Time |
| --------------------- | ---------- | ---------- | ---------- | ------------- |
| Baseline CNN          | 0.1354     | 0.1299     | 0.7557     | ~20–25 min    |
| ResNet18 (Transfer)   | 0.8727     | 0.8704     | 0.9976     | 518 sec       |
| ResNet18 (Fine-Tuned) | **0.9224** | **0.9219** | **0.9988** | 572 sec       |


The comparison results show a clear progression in performance from the baseline CNN to transfer learning and then fine-tuning. While the baseline CNN produced weak results, ResNet18 greatly improved accuracy, F1-score, and AUC, and the fine-tuned version achieved the best overall performance, confirming that additional adaptation of pretrained features was beneficial for this dataset.

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
from sklearn.preprocessing import label_binarize
import torch.nn.functional as F
import numpy as np

efficientnet.eval()

all_preds = []
all_labels = []
all_probs = []

with torch.no_grad():
    for images, labels in val_loader:
        outputs = efficientnet(images)
        probs = F.softmax(outputs, dim=1)
        _, preds = torch.max(outputs, 1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())
        all_probs.extend(probs.cpu().numpy())

all_preds = np.array(all_preds)
all_labels = np.array(all_labels)
all_probs = np.array(all_probs)

acc = accuracy_score(all_labels, all_preds)
f1 = f1_score(all_labels, all_preds, average="macro")

all_labels_bin = label_binarize(all_labels, classes=np.arange(num_classes))
auc = roc_auc_score(all_labels_bin, all_probs, multi_class="ovr", average="macro")

print("EfficientNet Accuracy:", acc)
print("EfficientNet F1-score:", f1)
print("EfficientNet AUC:", auc)
```

EfficientNet-B0 achieved strong classification performance, showing that pretrained CNN features remained highly effective for this dataset. Its results can be directly compared with ResNet18 to examine whether a more efficient architecture provides similar or better accuracy, F1-score, and AUC.

## 3.1. Data Augmentation Study

This subsection investigates whether stronger augmentation strategies can improve model generalization beyond the standard preprocessing pipeline. The models are compared under different augmentation settings to measure their impact

```python
from torchvision import transforms
from torch.utils.data import DataLoader

# AUGMENTED transform for TRAIN only
aug_train_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(15),
    transforms.ToTensor(),
])

# NO augmentation for validation/test
eval_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

# New datasets for augmentation study
aug_train_dataset = PetDataset(train_df, transform=aug_train_transform)
aug_val_dataset = PetDataset(val_df, transform=eval_transform)
aug_test_dataset = PetDataset(test_df, transform=eval_transform)

# New loaders for augmentation study
aug_train_loader = DataLoader(aug_train_dataset, batch_size=32, shuffle=True)
aug_val_loader = DataLoader(aug_val_dataset, batch_size=32, shuffle=False)
aug_test_loader = DataLoader(aug_test_dataset, batch_size=32, shuffle=False)

print("Aug train size:", len(aug_train_dataset))
print("Aug val size:", len(aug_val_dataset))
print("Aug test size:", len(aug_test_dataset))
```

```python
import torch
import torch.nn as nn
from torchvision import models

# device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# number of classes
num_classes = len(train_df["label"].unique())
print("Number of classes:", num_classes)

# fresh EfficientNet-B0 model
aug_effnet = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.DEFAULT)

# replace classifier
in_features = aug_effnet.classifier[1].in_features
aug_effnet.classifier[1] = nn.Linear(in_features, num_classes)

aug_effnet = aug_effnet.to(device)

# loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(aug_effnet.parameters(), lr=1e-3)

print("Augmentation EfficientNet-B0 is ready.")
```

```python
import time
import copy
import torch

num_epochs = 2  # change later if needed

best_val_acc = 0.0
best_model_wts = copy.deepcopy(aug_effnet.state_dict())

aug_train_losses = []
aug_val_accuracies = []

start_time = time.time()

for epoch in range(num_epochs):
    # -------------------
    # Training
    # -------------------
    aug_effnet.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in aug_train_loader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = aug_effnet(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)

        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    epoch_train_loss = running_loss / total
    epoch_train_acc = correct / total
    aug_train_losses.append(epoch_train_loss)

    # -------------------
    # Validation
    # -------------------
    aug_effnet.eval()
    val_correct = 0
    val_total = 0

    with torch.no_grad():
        for images, labels in aug_val_loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = aug_effnet(images)
            _, preds = torch.max(outputs, 1)

            val_correct += (preds == labels).sum().item()
            val_total += labels.size(0)

    epoch_val_acc = val_correct / val_total
    aug_val_accuracies.append(epoch_val_acc)

    # save best model
    if epoch_val_acc > best_val_acc:
        best_val_acc = epoch_val_acc
        best_model_wts = copy.deepcopy(aug_effnet.state_dict())

    print(f"[Aug] Epoch {epoch+1}/{num_epochs} - Loss: {epoch_train_loss:.4f} - Train Acc: {epoch_train_acc:.4f} - Val Acc: {epoch_val_acc:.4f}")

end_time = time.time()
aug_training_time = end_time - start_time

# load best model weights
aug_effnet.load_state_dict(best_model_wts)

print(f"\nBest Val Accuracy: {best_val_acc:.4f}")
print(f"Aug training time: {aug_training_time:.1f} seconds")
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

aug_effnet.eval()

all_preds = []
all_labels = []
all_probs = []

with torch.no_grad():
    for images, labels in aug_test_loader:
        images = images.to(device)
        labels = labels.to(device)

        outputs = aug_effnet(images)
        probs = torch.softmax(outputs, dim=1)
        preds = torch.argmax(probs, dim=1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())
        all_probs.extend(probs.cpu().numpy())

all_preds = np.array(all_preds)
all_labels = np.array(all_labels)
all_probs = np.array(all_probs)

aug_accuracy = accuracy_score(all_labels, all_preds)
aug_f1 = f1_score(all_labels, all_preds, average="weighted")
aug_auc = roc_auc_score(all_labels, all_probs, multi_class="ovr")

print("Aug Accuracy:", aug_accuracy)
print("Aug F1-score:", aug_f1)
print("Aug AUC:", aug_auc)
```

## 3.2. Mixed-based Augmentation

This subsection explores mixed-based augmentation by combining training samples during learning rather than applying only geometric transformations

```python
import numpy as np
import torch

def mixup_data(x, y, alpha=0.4):
    """
    Returns mixed inputs, pairs of targets, and lambda
    """
    if alpha > 0:
        lam = np.random.beta(alpha, alpha)
    else:
        lam = 1.0

    batch_size = x.size(0)
    index = torch.randperm(batch_size).to(x.device)

    mixed_x = lam * x + (1 - lam) * x[index, :]
    y_a, y_b = y, y[index]

    return mixed_x, y_a, y_b, lam


def mixup_criterion(criterion, pred, y_a, y_b, lam):
    return lam * criterion(pred, y_a) + (1 - lam) * criterion(pred, y_b)
```

```python
import torch
import torch.nn as nn
from torchvision import models

# device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# number of classes
num_classes = len(train_df["label"].unique())
print("Number of classes:", num_classes)

# fresh EfficientNet-B0 for Mixup
mixup_effnet = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.DEFAULT)

# replace classifier
in_features = mixup_effnet.classifier[1].in_features
mixup_effnet.classifier[1] = nn.Linear(in_features, num_classes)

mixup_effnet = mixup_effnet.to(device)

# loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(mixup_effnet.parameters(), lr=1e-3)

print("Mixup EfficientNet-B0 is ready.")
```

```python
import time
import copy
import torch

num_epochs = 2
alpha = 0.4

best_val_acc = 0.0
best_model_wts = copy.deepcopy(mixup_effnet.state_dict())

mixup_train_losses = []
mixup_val_accuracies = []

start_time = time.time()
```

```python
for epoch in range(num_epochs):
    mixup_effnet.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images = images.to(device)
        labels = labels.to(device)

        mixed_images, targets_a, targets_b, lam = mixup_data(images, labels, alpha=alpha)

        optimizer.zero_grad()
        outputs = mixup_effnet(mixed_images)
        loss = mixup_criterion(criterion, outputs, targets_a, targets_b, lam)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)

        preds = torch.argmax(outputs, dim=1)
        correct += (
            lam * (preds == targets_a).sum().item()
            + (1 - lam) * (preds == targets_b).sum().item()
        )
        total += labels.size(0)

    epoch_train_loss = running_loss / total
    epoch_train_acc = correct / total
    mixup_train_losses.append(epoch_train_loss)

    mixup_effnet.eval()
    val_correct = 0
    val_total = 0

    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = mixup_effnet(images)
            preds = torch.argmax(outputs, dim=1)

            val_correct += (preds == labels).sum().item()
            val_total += labels.size(0)

    epoch_val_acc = val_correct / val_total
    mixup_val_accuracies.append(epoch_val_acc)

    if epoch_val_acc > best_val_acc:
        best_val_acc = epoch_val_acc
        best_model_wts = copy.deepcopy(mixup_effnet.state_dict())

    print(f"[Mixup] Epoch {epoch+1}/{num_epochs} - Loss: {epoch_train_loss:.4f} - Train Acc: {epoch_train_acc:.4f} - Val Acc: {epoch_val_acc:.4f}")
```

```python
end_time = time.time()
mixup_training_time = end_time - start_time

mixup_effnet.load_state_dict(best_model_wts)

print(f"Best Val Accuracy: {best_val_acc:.4f}")
print(f"Mixup training time: {mixup_training_time:.1f} seconds")
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

mixup_effnet.eval()

all_preds = []
all_labels = []
all_probs = []

with torch.no_grad():
    for images, labels in test_loader:
        images = images.to(device)
        labels = labels.to(device)

        outputs = mixup_effnet(images)
        probs = torch.softmax(outputs, dim=1)
        preds = torch.argmax(probs, dim=1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())
        all_probs.extend(probs.cpu().numpy())

all_preds = np.array(all_preds)
all_labels = np.array(all_labels)
all_probs = np.array(all_probs)

mixup_accuracy = accuracy_score(all_labels, all_preds)
mixup_f1 = f1_score(all_labels, all_preds, average="weighted")
mixup_auc = roc_auc_score(all_labels, all_probs, multi_class="ovr")

print("Mixup Accuracy:", mixup_accuracy)
print("Mixup F1-score:", mixup_f1)
print("Mixup AUC:", mixup_auc)
```

### 3.3.3. Hybrid Model

The hybrid model uses EfficientNet-B0 as a pretrained CNN backbone and replaces its original classifier with a custom multilayer perceptron head. This design was intended to combine strong pretrained visual feature extraction with a more flexible task-specific classifier.

```python
import torch
import torch.nn as nn
from torchvision import models

# device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# number of classes
num_classes = len(train_df["label"].unique())
print("Number of classes:", num_classes)

# load pretrained EfficientNet-B0
hybrid_model = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.DEFAULT)

# get input features of original classifier
in_features = hybrid_model.classifier[1].in_features

# replace classifier with hybrid MLP head
hybrid_model.classifier = nn.Sequential(
    nn.Dropout(0.3),
    nn.Linear(in_features, 256),
    nn.ReLU(),
    nn.Dropout(0.3),
    nn.Linear(256, num_classes)
)

hybrid_model = hybrid_model.to(device)

# loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(hybrid_model.parameters(), lr=1e-3)

print("Hybrid CNN + MLP model is ready.")
```

```python
import time
import copy
import torch

num_epochs = 2

best_val_acc = 0.0
best_model_wts = copy.deepcopy(hybrid_model.state_dict())

hybrid_train_losses = []
hybrid_val_accuracies = []

start_time = time.time()
```

```python
for epoch in range(num_epochs):
    hybrid_model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = hybrid_model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)

        preds = torch.argmax(outputs, dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    epoch_train_loss = running_loss / total
    epoch_train_acc = correct / total
    hybrid_train_losses.append(epoch_train_loss)

    hybrid_model.eval()
    val_correct = 0
    val_total = 0

    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = hybrid_model(images)
            preds = torch.argmax(outputs, dim=1)

            val_correct += (preds == labels).sum().item()
            val_total += labels.size(0)

    epoch_val_acc = val_correct / val_total
    hybrid_val_accuracies.append(epoch_val_acc)

    if epoch_val_acc > best_val_acc:
        best_val_acc = epoch_val_acc
        best_model_wts = copy.deepcopy(hybrid_model.state_dict())

    print(f"[Hybrid] Epoch {epoch+1}/{num_epochs} - Loss: {epoch_train_loss:.4f} - Train Acc: {epoch_train_acc:.4f} - Val Acc: {epoch_val_acc:.4f}")
```

```python
end_time = time.time()
hybrid_training_time = end_time - start_time

hybrid_model.load_state_dict(best_model_wts)

print(f"Best Val Accuracy: {best_val_acc:.4f}")
print(f"Hybrid training time: {hybrid_training_time:.1f} seconds")
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

hybrid_model.eval()

all_preds = []
all_labels = []
all_probs = []

with torch.no_grad():
    for images, labels in test_loader:
        images = images.to(device)
        labels = labels.to(device)

        outputs = hybrid_model(images)
        probs = torch.softmax(outputs, dim=1)
        preds = torch.argmax(probs, dim=1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())
        all_probs.extend(probs.cpu().numpy())

all_preds = np.array(all_preds)
all_labels = np.array(all_labels)
all_probs = np.array(all_probs)

hybrid_accuracy = accuracy_score(all_labels, all_preds)
hybrid_f1 = f1_score(all_labels, all_preds, average="weighted")
hybrid_auc = roc_auc_score(all_labels, all_probs, multi_class="ovr")

print("Hybrid Accuracy:", hybrid_accuracy)
print("Hybrid F1-score:", hybrid_f1)
print("Hybrid AUC:", hybrid_auc)
```

## 3.5. Class imbalance

This section examines techniques used to reduce the effect of uneven class representation during training. Although the dataset was relatively balanced, these methods were still tested to determine whether they could further improve learning stability and class-wise performance

### 3.5.1 Weighted Sampling

Weighted sampling was applied to give relatively more importance to underrepresented classes during training. This helps the model see a more balanced stream of examples, even when the original class counts are not perfectly equal

```python
train_df["label"].value_counts().sort_values()
```

```python
import torch
from torch.utils.data import WeightedRandomSampler

# count samples per class
class_counts = train_df["label"].value_counts().sort_index()

# map each training label to its class weight
class_weights = 1.0 / class_counts
sample_weights = train_df["label"].map(class_weights).values

sample_weights = torch.DoubleTensor(sample_weights)

weighted_sampler = WeightedRandomSampler(
    weights=sample_weights,
    num_samples=len(sample_weights),
    replacement=True
)

print("Weighted sampler ready.")
```

```python
from torch.utils.data import DataLoader

weighted_train_loader = DataLoader(
    train_dataset,
    batch_size=32,
    sampler=weighted_sampler
)

print("Weighted train loader ready.")
```

```python
import torch
import torch.nn as nn
from torchvision import models

# device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# number of classes
num_classes = len(train_df["label"].unique())
print("Number of classes:", num_classes)

# fresh EfficientNet-B0
weighted_effnet = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.DEFAULT)

# replace classifier
in_features = weighted_effnet.classifier[1].in_features
weighted_effnet.classifier[1] = nn.Linear(in_features, num_classes)

weighted_effnet = weighted_effnet.to(device)

# loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(weighted_effnet.parameters(), lr=1e-3)

print("Weighted-sampling EfficientNet is ready.")
```

```python
import time
import copy
import torch

num_epochs = 2

best_val_acc = 0.0
best_model_wts = copy.deepcopy(weighted_effnet.state_dict())

weighted_train_losses = []
weighted_val_accuracies = []

start_time = time.time()
```

```python
for epoch in range(num_epochs):
    weighted_effnet.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in weighted_train_loader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = weighted_effnet(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)

        preds = torch.argmax(outputs, dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    epoch_train_loss = running_loss / total
    epoch_train_acc = correct / total
    weighted_train_losses.append(epoch_train_loss)

    weighted_effnet.eval()
    val_correct = 0
    val_total = 0

    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = weighted_effnet(images)
            preds = torch.argmax(outputs, dim=1)

            val_correct += (preds == labels).sum().item()
            val_total += labels.size(0)

    epoch_val_acc = val_correct / val_total
    weighted_val_accuracies.append(epoch_val_acc)

    if epoch_val_acc > best_val_acc:
        best_val_acc = epoch_val_acc
        best_model_wts = copy.deepcopy(weighted_effnet.state_dict())

    print(f"[Weighted] Epoch {epoch+1}/{num_epochs} - Loss: {epoch_train_loss:.4f} - Train Acc: {epoch_train_acc:.4f} - Val Acc: {epoch_val_acc:.4f}")
```

```python
end_time = time.time()
weighted_training_time = end_time - start_time

weighted_effnet.load_state_dict(best_model_wts)

print(f"Best Val Accuracy: {best_val_acc:.4f}")
print(f"Weighted EfficientNet training time: {weighted_training_time:.1f} seconds")
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

weighted_effnet.eval()

all_preds = []
all_labels = []
all_probs = []

with torch.no_grad():
    for images, labels in test_loader:
        images = images.to(device)
        labels = labels.to(device)

        outputs = weighted_effnet(images)
        probs = torch.softmax(outputs, dim=1)
        preds = torch.argmax(probs, dim=1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())
        all_probs.extend(probs.cpu().numpy())

all_preds = np.array(all_preds)
all_labels = np.array(all_labels)
all_probs = np.array(all_probs)

weighted_accuracy = accuracy_score(all_labels, all_preds)
weighted_f1 = f1_score(all_labels, all_preds, average="weighted")
weighted_auc = roc_auc_score(all_labels, all_probs, multi_class="ovr")

print("Weighted EfficientNet Accuracy:", weighted_accuracy)
print("Weighted EfficientNet F1-score:", weighted_f1)
print("Weighted EfficientNet AUC:", weighted_auc)
```

### 3.5.2. Focal Loss

Focal loss was tested as an alternative imbalance-handling method that places more emphasis on hard-to-classify examples during training. This can help the model focus less on already easy predictions and more on samples that contribute most to classification errors.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class FocalLoss(nn.Module):
    def __init__(self, alpha=1.0, gamma=2.0, reduction="mean"):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction

    def forward(self, inputs, targets):
        ce_loss = F.cross_entropy(inputs, targets, reduction="none")
        pt = torch.exp(-ce_loss)
        focal_loss = self.alpha * (1 - pt) ** self.gamma * ce_loss

        if self.reduction == "mean":
            return focal_loss.mean()
        elif self.reduction == "sum":
            return focal_loss.sum()
        else:
            return focal_loss
```

```python
import torch
import torch.nn as nn
from torchvision import models

# device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# number of classes
num_classes = len(train_df["label"].unique())
print("Number of classes:", num_classes)

# fresh EfficientNet-B0
focal_effnet = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.DEFAULT)

# replace classifier
in_features = focal_effnet.classifier[1].in_features
focal_effnet.classifier[1] = nn.Linear(in_features, num_classes)

focal_effnet = focal_effnet.to(device)

# focal loss + optimizer
criterion = FocalLoss(alpha=1.0, gamma=2.0)
optimizer = torch.optim.Adam(focal_effnet.parameters(), lr=1e-3)

print("Focal-loss EfficientNet is ready.")
```

```python
import time
import copy
import torch

num_epochs = 2

best_val_acc = 0.0
best_model_wts = copy.deepcopy(focal_effnet.state_dict())

focal_train_losses = []
focal_val_accuracies = []

start_time = time.time()
```

```python
for epoch in range(num_epochs):
    focal_effnet.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = focal_effnet(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)

        preds = torch.argmax(outputs, dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    epoch_train_loss = running_loss / total
    epoch_train_acc = correct / total
    focal_train_losses.append(epoch_train_loss)

    focal_effnet.eval()
    val_correct = 0
    val_total = 0

    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = focal_effnet(images)
            preds = torch.argmax(outputs, dim=1)

            val_correct += (preds == labels).sum().item()
            val_total += labels.size(0)

    epoch_val_acc = val_correct / val_total
    focal_val_accuracies.append(epoch_val_acc)

    if epoch_val_acc > best_val_acc:
        best_val_acc = epoch_val_acc
        best_model_wts = copy.deepcopy(focal_effnet.state_dict())

    print(f"[Focal] Epoch {epoch+1}/{num_epochs} - Loss: {epoch_train_loss:.4f} - Train Acc: {epoch_train_acc:.4f} - Val Acc: {epoch_val_acc:.4f}")
```

```python
end_time = time.time()
focal_training_time = end_time - start_time

focal_effnet.load_state_dict(best_model_wts)

print(f"Best Val Accuracy: {best_val_acc:.4f}")
print(f"Focal EfficientNet training time: {focal_training_time:.1f} seconds")
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

focal_effnet.eval()

all_preds = []
all_labels = []
all_probs = []

with torch.no_grad():
    for images, labels in test_loader:
        images = images.to(device)
        labels = labels.to(device)

        outputs = focal_effnet(images)
        probs = torch.softmax(outputs, dim=1)
        preds = torch.argmax(probs, dim=1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())
        all_probs.extend(probs.cpu().numpy())

all_preds = np.array(all_preds)
all_labels = np.array(all_labels)
all_probs = np.array(all_probs)

focal_accuracy = accuracy_score(all_labels, all_preds)
focal_f1 = f1_score(all_labels, all_preds, average="weighted")
focal_auc = roc_auc_score(all_labels, all_probs, multi_class="ovr")

print("Focal EfficientNet Accuracy:", focal_accuracy)
print("Focal EfficientNet F1-score:", focal_f1)
print("Focal EfficientNet AUC:", focal_auc)
```

```python
import numpy as np
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report
)

focal_effnet.eval()

all_preds = []
all_labels = []
all_probs = []

with torch.no_grad():
    for images, labels in test_loader:
        images = images.to(device)
        labels = labels.to(device)

        outputs = focal_effnet(images)
        probs = torch.softmax(outputs, dim=1)
        preds = torch.argmax(probs, dim=1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())
        all_probs.extend(probs.cpu().numpy())

all_preds = np.array(all_preds)
all_labels = np.array(all_labels)
all_probs = np.array(all_probs)

final_accuracy = accuracy_score(all_labels, all_preds)
final_precision = precision_score(all_labels, all_preds, average="weighted")
final_recall = recall_score(all_labels, all_preds, average="weighted")
final_f1 = f1_score(all_labels, all_preds, average="weighted")
final_auc = roc_auc_score(all_labels, all_probs, multi_class="ovr")

final_cm = confusion_matrix(all_labels, all_preds)
final_report = classification_report(all_labels, all_preds)

print("Accuracy:", final_accuracy)
print("Precision:", final_precision)
print("Recall:", final_recall)
print("F1-score:", final_f1)
print("ROC-AUC:", final_auc)

print("\nConfusion Matrix:\n", final_cm)
print("\nClassification Report:\n", final_report)
```
